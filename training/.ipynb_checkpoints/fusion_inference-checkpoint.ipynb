```python
import os
os.environ["CUDA_VISIBLE_DEVICES"] = "-1"
import numpy as np
import librosa
import tensorflow as tf

if os.path.basename(os.getcwd()) == "training":
    PROJECT_ROOT = os.path.dirname(os.getcwd())
else:
    PROJECT_ROOT = os.getcwd()

genre_model = tf.keras.models.load_model(
    os.path.join(PROJECT_ROOT, "models", "genre_mel_cnn.keras")
)

emotion_model = tf.keras.models.load_model(
    os.path.join(PROJECT_ROOT, "models", "emotion_hybrid_model.keras")
)

print("Models loaded successfully.")
```

```python
SAMPLE_RATE = 22050
N_MELS = 128
SEGMENT_DURATION = 3

def extract_mel_for_inference(file_path):
    signal, sr = librosa.load(file_path, sr=SAMPLE_RATE)

    mel = librosa.feature.melspectrogram(
        y=signal[:SAMPLE_RATE * SEGMENT_DURATION],
        sr=sr,
        n_mels=N_MELS
    )

    mel = librosa.power_to_db(mel)

    mel = (mel - np.mean(mel)) / np.std(mel)

    return mel[np.newaxis, ..., np.newaxis]
```

```python
def extract_stats(segment, sr):
    tempo, _ = librosa.beat.beat_track(y=segment, sr=sr)
    tempo = float(np.squeeze(tempo))  # Ensure scalar
    centroid = float(np.mean(librosa.feature.spectral_centroid(y=segment, sr=sr)))
    rms = float(np.mean(librosa.feature.rms(y=segment)))
    zcr = float(np.mean(librosa.feature.zero_crossing_rate(segment)))
    return np.array([[tempo, centroid, rms, zcr]])
```

```python
file_path = "../test_audio/Classical.wav"

signal, sr = librosa.load(file_path, sr=SAMPLE_RATE)

segment = signal[:SAMPLE_RATE * SEGMENT_DURATION]

mel = extract_mel_for_inference(file_path)
stats = extract_stats(segment, sr)

genre_probs = genre_model.predict(mel)[0]
valence, arousal = emotion_model.predict([mel, stats])[0]

print("Genre probabilities:", genre_probs)
print("Valence:", valence)
print("Arousal:", arousal)
```
